```python
fall.Date = pd.to_datetime(fall.Date,format='%y%m%d')
```

```python
fall.Month = pd.to_datetime(fall.Month, format='%b%y')
```

```python
coffee_dec_13_17 = coffee_dec.loc[(coffee_dec.Product == 'Coffee "C"(ICE)')
                   & (coffee_dec.Month.dt.month == 12)
                   & (coffee_dec.Date >= '2013-01-01')
                   & (coffee_dec.Date <= '2017-12-31')
                   & (coffee_dec.Date.dt.year == coffee_dec.Month.dt.year)]
```
